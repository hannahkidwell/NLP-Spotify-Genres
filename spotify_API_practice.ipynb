# Create a DataFrame

```python
# Create a dictionary of categories and playlists
category_playlist_ids = {}
for cat in categories_list:
    query_url = base_url + f'browse/categories/{cat}/playlists?limit=5'
    playlists = []
    results = requests.get(query_url, headers={
        'Content-Type': 'application/json',
        'Authorization': f'Bearer {spotify_token}',
        'country': 'US'}).json()
    for id, playlist in enumerate(results['playlists']['items']):
        playlists.append(playlist['id'])
    category_playlist_ids[cat] = playlists
category_playlist_ids
```

```python
# Create a DataFrame of song names, artists, and categories
song_artist_category_df = pd.DataFrame(columns=['song', 'artist', 'category'])
index = 0
for category in category_playlist_ids:
    songs_list = []
    for playlist_id in category_playlist_ids[category]:
        query_url = base_url + f'playlists/{playlist_id}/tracks?limit=5'
        results = requests.get(query_url, headers={
            'Content-Type': 'application/json',
            'Authorization': f'Bearer {spotify_token}',
            'country': 'US'}).json()
        for id, item in enumerate(results['items']):
            song = item['track']['name']
            artist = item['track']['artists'][0]['name']
            song_artist_category_df.loc[index] = [song, artist, category]
            index+=1
song_artist_category_df.head()
```

```python
# Save DataFrame to CSV
song_artist_category_df.to_csv('song_artist_category.csv', index=False)
```

```python
# Create dictionary of categories and songs
category_songs = {}
for category in category_playlist_ids:
    songs_list = []
    for playlist_id in category_playlist_ids[category]:
        query_url = base_url + f'playlists/{playlist_id}/tracks?limit=5'
        results = requests.get(query_url, headers={
            'Content-Type': 'application/json',
            'Authorization': f'Bearer {spotify_token}',
            'country': 'US'}).json()
        for id, item in enumerate(results['items']):
            songs_list.append(item['track']['name'])
    category_songs[category] = songs_list
```

```python
# Add NaNs to columns with less than 25 values
for category in category_songs:
    length = len(category_songs[category])
    if length == 0:
        print(category)
    elif length != 25:
        new_list = category_songs[category]
        nans = 25-length
        for i in range(nans):
            new_list.append(np.nan)
        category_songs[category] = new_list
```

```python
# Remove the in_the_car category with no songs
category_songs.pop('in_the_car')
```

```python
# Create DataFrame of songs and categories
songs_by_category = pd.DataFrame.from_dict(category_songs)
songs_by_category
```

```python
# Save DataFrame to CSV
songs_by_category.to_csv('songs_by_category.csv', index=False)
```
