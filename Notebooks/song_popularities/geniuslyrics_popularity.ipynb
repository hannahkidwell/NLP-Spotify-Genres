```python
# Import dependencies
import pandas as pd
import numpy as np
import re
import time
```

```python
# Create DataFrame from Spotify API calls
song_genres_popularity = pd.read_csv('../Data/song_genres_popularity.csv')
song_lyrics_df = pd.read_csv('../Data/song_lyrics.csv')
song_genres_popularity.head(3)
```

```python
# Add popularity column to song_lyrics_df
t0 = time.time()
popular_dict = {}
for index, row in song_genres_popularity.iterrows():
    popular_dict[row['song_id']] = row['popularity']
popularities = []
for index, row in song_lyrics_df.iterrows():
    popularities.append(popular_dict[row['song_id']])
t1 = time.time()
print(f'Run time: {t1-t0} seconds')
song_lyrics_df.insert(5, 'popularity', popularities)
song_lyrics_df.head(3)
```

```python
# Save DataFrame to CSV
song_lyrics_df.to_csv('../Data/song_lyrics_popularity.csv', index=False)
```
