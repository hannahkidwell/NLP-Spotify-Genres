```python
import lyricsgenius
from config import genius_token
import pandas as pd
import numpy as np
import re
import time
```

```python
genius = lyricsgenius.Genius(genius_token)
```

```python
# Create DataFrame from Spotify API calls
song_artist_category_df = pd.read_csv('../../preliminary_dataframes/song_artist_category3.csv')
song_artist_category_df
```

```python
# Time the code that creates a list of lyrics for the first 100 songs
t0 = time.time()
# Create a list of lyrics for the first 100 songs
lyrics_list = []
for index, row in song_artist_category_df[0:100].iterrows():
    print(index)
    song = row['song']
    artist = row['artist']
    results = genius.search_songs(f'{song} {artist}')
    try:
        found_song = False
        for id, hit in enumerate(results['hits']):
            artist_name = hit['result']['primary_artist']['name'].lower()
            if row['artist'].lower() in artist_name:
                found_song = True
                song_id = hit['result']['id']
                lyrics = genius.lyrics(urlthing=song_id)
                lyrics_list.append(lyrics)
                break
        if found_song == False:
            lyrics_list.append(None)
            print('Song not found.')
    except:
        lyrics_list.append(None)
        print('No result found.')
t1 = time.time()
print(t1-t0)
len(lyrics_list)
```

```python
# Clean the lyrics column of the DataFrame
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('\u205f', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('  ', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('  ', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('\n', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('-', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('?', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('!', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(',', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('(', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(')', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace("'", '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('’', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('\[(.*?)\]', '', regex=True)
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(',', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('.', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('`', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('[', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(']', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('{', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('}', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace("'", '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('"', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('”', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(':', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(';', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('*', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('“', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('_', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('—', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('–', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('\\u2005', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('¨', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('…', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('‘', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('=', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('·', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('е', 'e')
song_artist_category_df.head()
```

```python
# Count non alpha words
non_alpha_words = []
for index, row in song_artist_category_df.iterrows():
    filtered_words = row['lyrics']
    count = 0
    for word in list(set(str(filtered_words).strip().split(' '))):
        if not re.match(r'[a-zA-Z0-9/#$&+]+$', word):
            count+=1
    non_alpha_words.append(count)
len(non_alpha_words)
```

```python
# Add a column for alpha word count
song_artist_category_df['non_alpha_words'] = non_alpha_words
song_artist_category_df[song_artist_category_df['non_alpha_words']>10][['song', 'artist', 'lyrics']]
```

```python
# Print non alpha words for each song
for item, row in song_artist_category_df[song_artist_category_df['non_alpha_words']>10][['song', 'artist', 'lyrics']].iterrows():
    filtered_words = row['lyrics']
    for word in list(set(str(filtered_words).strip().split(' '))):
        if not re.match(r'[a-zA-Z0-9,.\/#$&+]+$', word):
            print(row['song'], word)
```

```python
# Remove songs with more than 10 non alpha words
song_artist_category_df = song_artist_category_df[song_artist_category_df['non_alpha_words']<=10]
song_artist_category_df
```
