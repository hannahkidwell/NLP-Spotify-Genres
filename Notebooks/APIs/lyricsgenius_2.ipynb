```python
# Clean the lyrics column of the DataFrame
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('\u205f', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('  ', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('  ', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('\n', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('-', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('?', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('!', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(',', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('(', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(')', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace("'", '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('’', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('\[(.*?)\]', '', regex=True)
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(',', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('.', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('`', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('[', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(']', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('{', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('}', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace("'", '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('"', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('”', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(':', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace(';', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('*', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('“', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('_', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('—', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('–', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('\\u2005', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('¨', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('…', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('‘', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('=', ' ')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('·', '')
song_artist_category_df['lyrics'] = song_artist_category_df['lyrics'].str.replace('е', 'e')
song_artist_category_df.head()
```

```python
# Count non alpha words
non_alpha_words = []
for index, row in song_artist_category_df.iterrows():
    filtered_words = row['lyrics']
    count = 0
    for word in list(set(str(filtered_words).strip().split(' '))):
        if not re.match(r'[a-zA-Z0-9,./#$&+]+$', word):
            count+=1
    non_alpha_words.append(count)
len(non_alpha_words)
```

```python
# Add a column for alpha word count
song_artist_category_df['non_alpha_words'] = non_alpha_words
song_artist_category_df[song_artist_category_df['non_alpha_words']>10][['song', 'artist', 'lyrics']]
```
